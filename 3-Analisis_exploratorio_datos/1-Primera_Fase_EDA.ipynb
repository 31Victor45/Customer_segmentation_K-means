```python
# Cargamos las librerias con las que trabajaremos
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Vista inicial de los datos
df_customer = pd.read_csv("dataset/new_shopping_trends.csv")
df_customer.head()
```

Obtengamos las estadisticas descriptivas de nuestros datos usando el metodo .describe()

```python
df_customer.describe()
```

Una ves obtenido los estadisticos principales vamos a visualizar las distintas categorias que prensetan las variables categoricas del dataset asi como el pociento que representan cada una de las mismas

```python
# Identificamos las variables categóricas
categorical_cols = df_customer.select_dtypes(include=['object']).columns
total_count = len(df_customer)

print("# Visualización de variables categóricas con Barplots y Tablas")
for col in categorical_cols:
    # Barplot de Frecuencia Absoluta
    plt.figure(figsize=(10, 5))
    ax = sns.countplot(data=df_customer, y=col, order=df_customer[col].value_counts().index, palette='viridis')
    plt.title(f'Distribución de la variable: {col}', fontsize=15)
    plt.xlabel('Conteo', fontsize=12)
    plt.ylabel(col, fontsize=12)

    # Añadimos la cantidad (etiqueta) y el porcentaje en la cima de cada barra
    for p in ax.patches:
        width = p.get_width()
        percentage = (width / total_count) * 100
        plt.text(width + 10, p.get_y() + p.get_height() / 2,
                 f'{width:1.0f} ({percentage:.1f}%)',
                 ha='left', va='center', fontsize=10, color='black', weight='bold')
    plt.tight_layout()
    plt.show()
```

Vamos a indentificar si en las columnas numericas de nuestro dataframe existen valores atipicos (outlayers)

```python
# Identificamos las variables numéricas para la detección de outliers
numeric_cols = df_customer.select_dtypes(include=['int64', 'float64']).columns
print("# Detección de Outliers en variables numéricas con Boxplots")

# Creamos una lista de colores para los boxplots
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold', 'violet', 'lightsalmon']

# Creamos un boxplot para cada columna numérica
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols):
    plt.subplot(2, 3, i + 1)
    # Usamos el operador módulo para repetir los colores si hay más columnas que colores
    sns.boxplot(x=df_customer[col], color=colors[i % len(colors)])
    plt.title(f'Boxplot de {col}', fontsize=12)
    plt.xlabel(col, fontsize=10)
    plt.tight_layout()
plt.show()
```

```python
# Contamos outliers para cada variable numerica usando el método IQR
outliers_count = {}
for col in numeric_cols:
    Q1 = df_customer[col].quantile(0.25)
    Q3 = df_customer[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_customer[(df_customer[col] < lower_bound) | (df_customer[col] > upper_bound)]
    outliers_count[col] = outliers.shape[0]

print("Cantidad de outliers por variable numerica:")
for col, count in outliers_count.items():
    print(f"{col}: {count}")
```

Hasta ahora hemos analizado las variables categoricas de nuestro dataframe y tambien de las variables numericas conocemos sus estadisticas descriptivas y que estas no contienen valores atipicos pero nos falta por saber si existe correlacion entre estas que es lo que vamos a ver ahora

```python
# Calculamos la matriz de correlación
corr_matrix = df_customer[numeric_cols].corr()

# Visualizamos la matriz de correlación con un heatmap
plt.figure(figsize=(7, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriz de Correlación', fontsize=16)
plt.show()
```

Gracias a la matriz de correlacion anterior sabemos que no existe correlacion alguna entre las variables numericas de nuestro dataset
